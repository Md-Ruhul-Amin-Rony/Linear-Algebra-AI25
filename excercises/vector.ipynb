## Problem 11
For problems 1‚Äì10, now also compute the **length** (magnitude) of the resulting vectors.

**Reminder:** The length of a vector $\vec{v} = \begin{bmatrix} v_1 \\ v_2 \\ \vdots \\ v_n \end{bmatrix}$ is:

$$\|\vec{v}\| = \sqrt{v_1^2 + v_2^2 + \cdots + v_n^2}$$

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Function to plot 2D vectors
def plot_vectors_2d(vectors, labels, colors, title):
    """
    Plot multiple 2D vectors from origin
    vectors: list of 2D numpy arrays
    labels: list of strings for legend
    colors: list of color strings
    """
    plt.figure(figsize=(10, 8))
    plt.axhline(y=0, color='k', linewidth=0.5)
    plt.axvline(x=0, color='k', linewidth=0.5)
    plt.grid(True, alpha=0.3)
    
    for vec, label, color in zip(vectors, labels, colors):
        plt.arrow(0, 0, vec[0], vec[1], 
                 head_width=0.3, head_length=0.3, 
                 fc=color, ec=color, linewidth=2, 
                 length_includes_head=True, label=label)
        # Add text label at the end of arrow
        plt.text(vec[0]*1.1, vec[1]*1.1, label, fontsize=10, fontweight='bold')
    
    plt.xlabel('X', fontsize=12)
    plt.ylabel('Y', fontsize=12)
    plt.title(title, fontsize=14, fontweight='bold')
    plt.legend(fontsize=10)
    plt.axis('equal')
    plt.tight_layout()
    plt.show()

# Function to calculate vector magnitude
def magnitude(vec):
    """Calculate the magnitude (length) of a vector"""
    return np.sqrt(np.sum(vec**2))

print("‚úÖ Libraries imported and helper functions defined!")
```

## üìù Solution to Problem 1

**Given:** $\vec{a} = \begin{bmatrix} 2 \\ -1 \end{bmatrix}$, $\vec{b} = \begin{bmatrix} -3 \\ 4 \end{bmatrix}$, $k = 2$

**Find:** $k\vec{a} + 3\vec{b}$

**Step-by-step:**
1. First, multiply $\vec{a}$ by scalar $k = 2$: $k\vec{a} = 2\begin{bmatrix} 2 \\ -1 \end{bmatrix} = \begin{bmatrix} 4 \\ -2 \end{bmatrix}$
2. Next, multiply $\vec{b}$ by scalar $3$: $3\vec{b} = 3\begin{bmatrix} -3 \\ 4 \end{bmatrix} = \begin{bmatrix} -9 \\ 12 \end{bmatrix}$
3. Finally, add them: $k\vec{a} + 3\vec{b} = \begin{bmatrix} 4 \\ -2 \end{bmatrix} + \begin{bmatrix} -9 \\ 12 \end{bmatrix} = \begin{bmatrix} -5 \\ 10 \end{bmatrix}$

```python
# Problem 1 - Calculation
a = np.array([2, -1])
b = np.array([-3, 4])
k = 2

# Compute ka + 3b
ka = k * a
three_b = 3 * b
result_1 = ka + three_b

print("Problem 1:")
print(f"a = {a}")
print(f"b = {b}")
print(f"k = {k}")
print(f"\nka = {k} √ó {a} = {ka}")
print(f"3b = 3 √ó {b} = {three_b}")
print(f"\nResult: ka + 3b = {result_1}")
print(f"Magnitude: ||result|| = {magnitude(result_1):.4f}")

# Plot the vectors
plot_vectors_2d(
    [a, b, ka, three_b, result_1],
    ['a', 'b', 'ka', '3b', 'RESULT: ka+3b'],
    ['blue', 'green', 'lightblue', 'lightgreen', 'red'],
    'Problem 1: ka + 3b'
)
```

2. Multiply $\vec{f}$ by $5$: $5\vec{f} = 5\begin{bmatrix} -2 \\ 3 \end{bmatrix} = \begin{bmatrix} -10 \\ 15 \end{bmatrix}$

## üìù Solution to Problem 9 (3D Vector)

**Given:** $\vec{q} = \begin{bmatrix} u^2 \\ 4 \\ 3 \end{bmatrix}$, $\vec{r} = \begin{bmatrix} -3 \\ 0 \\ 2 \end{bmatrix}$, $u = 2$

**Find:** $4\vec{q} + \frac{1}{2}\vec{r} - \vec{r}$

**Step-by-step:**
1. Substitute $u = 2$: $\vec{q} = \begin{bmatrix} 4 \\ 4 \\ 3 \end{bmatrix}$
2. Multiply $\vec{q}$ by $4$: $4\vec{q} = 4\begin{bmatrix} 4 \\ 4 \\ 3 \end{bmatrix} = \begin{bmatrix} 16 \\ 16 \\ 12 \end{bmatrix}$
3. Simplify $\vec{r}$ term: $\frac{1}{2}\vec{r} - \vec{r} = -\frac{1}{2}\vec{r} = -\frac{1}{2}\begin{bmatrix} -3 \\ 0 \\ 2 \end{bmatrix} = \begin{bmatrix} 1.5 \\ 0 \\ -1 \end{bmatrix}$
4. Add: $4\vec{q} - \frac{1}{2}\vec{r} = \begin{bmatrix} 16 \\ 16 \\ 12 \end{bmatrix} + \begin{bmatrix} 1.5 \\ 0 \\ -1 \end{bmatrix} = \begin{bmatrix} 17.5 \\ 16 \\ 11 \end{bmatrix}$

```python
# Problem 9 - Calculation (3D Vector)
u = 2
q = np.array([u**2, 4, 3])
r = np.array([-3, 0, 2])

# Compute 4q + (1/2)r - r = 4q - (1/2)r
four_q = 4 * q
neg_half_r = -0.5 * r
result_9 = four_q + neg_half_r

print("Problem 9:")
print(f"u = {u}")
print(f"q = [u¬≤, 4, 3] = {q}")
print(f"r = {r}")
print(f"\n4q = 4 √ó {q} = {four_q}")
print(f"(1/2)r - r = -(1/2)r = -(1/2) √ó {r} = {neg_half_r}")
print(f"\nResult: 4q + (1/2)r - r = 4q - (1/2)r = {result_9}")
print(f"Magnitude: ||result|| = {magnitude(result_9):.4f}")
```

4. Subtract: $\frac{5}{2}\vec{s} - 3\vec{t} = \begin{bmatrix} 7.5 \\ -2.5 \\ -5 \end{bmatrix} - \begin{bmatrix} -6 \\ 15 \\ -6 \end{bmatrix} = \begin{bmatrix} 13.5 \\ -17.5 \\ 1 \end{bmatrix}$

## üìè Solution to Problem 11 (Magnitude Summary)

**Task:** Calculate the magnitude of all results from problems 1-10.

The magnitude (length) of a vector $\vec{v} = \begin{bmatrix} x \\ y \end{bmatrix}$ or $\vec{v} = \begin{bmatrix} x \\ y \\ z \end{bmatrix}$ is calculated using:

$$|\vec{v}| = \sqrt{x^2 + y^2} \quad \text{(2D)}$$
$$|\vec{v}| = \sqrt{x^2 + y^2 + z^2} \quad \text{(3D)}$$
